```python
import pandas as pd

df=pd.read_csv("hoja calculo pandas - Full 1 (2).csv")
df
```

- Calculeu quants objectes en òrbita són de: «Russian Federation», «China», «USA» i «ESA»
- Què és la ESA? Responeu-ho al mig del codi, amb una anotació.
- Calculeu quants satèl·lits estan en òrbita i quants ja no ho estan.

```python
russian = df["State"].str.contains("Russian Federation")
russian.value_counts()
russian[russian==True].count().sum()
```

```python
china = df["State"].str.contains("China")
china.value_counts()
china[china==True].count().sum()
```

```python
usa = df["State"].str.contains("USA")
usa.value_counts()
usa[usa==True].count().sum()
```

```python
esa = df["State"].str.contains("ESA")
esa.value_counts()
esa[esa==True].count().sum()
```
