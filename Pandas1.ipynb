```python
import pandas as pd

serie= pd.Series([12,58,63,47,81,35])
print(serie)

largada = (serie.size)
print(largada)

suma = serie.sum()
print("L’import total és:", suma)

media = serie.mean()
print(round(media,2))

raro = serie.value_counts()
print(raro)

max = serie.max()
print(max)

min = serie.min()
print(min)

raro_dos = serie.cumsum()
print(raro_dos)
```

```python
import pandas as pd

d= {'Persona': ["Pedro","Sonia","Luis","Marta","Julian","Nerea"],
    'Edad': [25,52,19,33,48,61],
    'Altura': [1.68,1.74,1.81,1.59,1.71,1.78]}
df = pd.DataFrame(data=d)
df
```
